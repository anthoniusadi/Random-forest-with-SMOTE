# Import library

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
```

```python
df=pd.read_csv("healthcare-dataset-stroke-data.csv")
df.head()
```

# Cleaning for missing value (drop)

```python
#check Nan data
nan_data = df[df.isna().any(axis=1)]
nan_data
```

```python
print(len(df))
df = df.dropna()
print(len(df))
df=df.drop(['id'],axis=1)
df.head()
```

# Visualize 

```python
sns.set(style="darkgrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 8))

sns.histplot(data=df,x='stroke',ax=axes[0,0],color='skyblue')
sns.histplot(data=df,x='gender',ax=axes[0,1],color='skyblue')

sns.histplot(data=df,x='age',ax=axes[1,0],color='skyblue')
sns.histplot(data=df,x='smoking_status',ax=axes[1,1],color='skyblue')

plt.show()
```

# Transform data

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
```

```python
#encode data categorial
df['gender']=encoder.fit_transform(df['gender'].values)
df['ever_married']=encoder.fit_transform(df['ever_married'].values)
df['work_type']=encoder.fit_transform(df['work_type'].values)
df['Residence_type']=encoder.fit_transform(df['Residence_type'].values)
df['smoking_status']=encoder.fit_transform(df['smoking_status'].values)
df.head()
```

```python
for col in df.columns:
    if df[col].max()>1:
        print(col)
```

```python
df[df['gender']>1]
```

```python
df.drop(df.index[df['gender']>1],inplace=True)
```

```python
df
```

# SCALING

```python
column_scale = [col for col in df.columns if df[col].max()>1]
scale = MinMaxScaler()
scaled = scale.fit_transform(df[column_scale])
scaled = pd.DataFrame(scaled,columns = column_scale)
# masukan pada kolom utama
for col in scaled:
    df[col]=scaled[col]
df = df.dropna()
df.head()
```

```python
sns.set(style="darkgrid")
fig, axes = plt.subplots(2, 2, figsize=(15, 8))

sns.histplot(data=df,x='stroke',ax=axes[0,0],color='skyblue')
sns.histplot(data=df,x='gender',ax=axes[0,1],color='skyblue')

sns.histplot(data=df,x='age',ax=axes[1,0],color='skyblue')
sns.histplot(data=df,x='smoking_status',ax=axes[1,1],color='skyblue')

plt.show()
```

# RANDOM FOREST WITHOUT SMOTE

```python
from sklearn.model_selection import train_test_split
x = df.drop('stroke',axis=1)
y = df['stroke']
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=50)
```

```python
from sklearn.ensemble import RandomForestClassifier as forest
from sklearn.metrics import accuracy_score, recall_score,confusion_matrix

#Train
model = forest(random_state=50)
model.fit(X_train,y_train)
pred = model.predict(X_test)
#Evaluate
print(f'Accuracy : {accuracy_score(y_test,pred)}, Recall : {recall_score(y_test,pred)}')
cm =confusion_matrix(y_test,pred)
plt.figure()
plt.title('Confusion Matrix')
sns.heatmap(cm,annot=True)
```

ATTENTION ON RECALL VALUE ! Model is unusable because there is imbalance data look into (stroke) graph 

# SMOTE for imbalance data

```python
from imblearn.over_sampling import SMOTE
print(f'Before SMOTE : {x.shape}')
sm = SMOTE(random_state=50)
x_sm,y_sm = sm.fit_resample(x,y)
print(f'After SMOTE : {x_sm.shape}')
print(f'Balance of positif & negatif class (%) : {y_sm.value_counts(normalize=True)*100} ')
```

# Visualize after SMOTE

```python
sns.set(style="darkgrid")
fig= plt.subplots(figsize=(8, 5))
sns.histplot(data=y_sm,color='skyblue')

plt.show()
```

# RANDOM forest using Data Smote

```python
x_train,x_test,y_train,y_test =train_test_split(x_sm,y_sm,test_size=0.25,random_state=50)
#Train random forest
model_smote = forest()
model_smote.fit(x_train,y_train)
pred_smote = model_smote.predict(x_test)
#Evaluate model
print(f'Accuracy : {accuracy_score(y_test,pred_smote)}, Recall_score : {recall_score(y_test,pred_smote)} ')
cm_smote =confusion_matrix(y_test,pred_smote)
plt.figure()
plt.title('Confusion Matrix')
sns.heatmap(cm_smote,annot=True,cmap='Blues')
```
